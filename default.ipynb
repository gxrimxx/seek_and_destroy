# Project Seek & Destroy

This is the official documentation and command reference for the Seek & Destroy ROS 2 project.

### ROSJect Customization
- Open your ROSJect
- Edit the default jupyter notebook file (`/home/user/notebook_ws/default.ipynb`)
- Wait for it to be saved automatically
- Your **Readme** will be updated based on your notebook

## 1. Initial Workspace Setup
Run the following commands to initialize the session environment:
```bash
cd ~/seek_destroy_ws/
./setup_session.sh
```

## 2. Standard Terminal Setup
Whenever you open a **new terminal**, always run these commands to source the workspace and link Gazebo:
```bash
cd ~/seek_destroy_ws/
source /opt/ros/humble/setup.bash
source install/setup.bash
export IGN_GAZEBO_SYSTEM_PLUGIN_PATH=/opt/ros/humble/lib/ign_ros2_control/system:$IGN_GAZEBO_SYSTEM_PLUGIN_PATH
```

## 3. Building & Manual Simulation
Use this to build the workspace and test the basic Gazebo simulation:
```bash
colcon build --symlink-install
ros2 launch rosbot_gazebo simulation.launch.py robot_model:=rosbot
```

## 4. Running SLAM (Mapping)
```bash
ros2 launch slam_toolbox online_async_launch.py \
  slam_params_file:=/home/user/seek_destroy_ws/my_slam_params.yaml \
  use_sim_time:=true
```

**RViz Setup:**
- Set **Fixed Frame** to `map`
- Click **Add** → By Topic → `/map` → Map
- Click **Add** → By Topic → `/scan` → LaserScan
- Click **Add** → RobotModel

## 5. Setting Nav2 Goals
```bash
ros2 launch nav2_bringup navigation_launch.py \
  use_sim_time:=true \
  params_file:=$HOME/seek_destroy_ws/src/seek_destroy/config/nav2_params.yaml
```

**RViz Setup:**
- **Add** → By Topic → `/local_costmap/costmap` → Map *(shows obstacle inflation)*
- **Add** → By Topic → `/global_costmap/costmap` → Map
- **Add** → By Topic → `/plan` → Path

## 6. Manual Explore Lite
Start Gazebo, SLAM, and Nav2 in three different terminals. Then, in a 4th terminal:
```bash
ros2 launch explore_lite explore.launch.py
```

## 7. Master Launch: Autonomous Exploration (Deliverable 2)
This single command automatically launches Gazebo, SLAM, Nav2, Explore Lite, RViz, and auto-saves the map!
```bash
cd ~/seek_destroy_ws
colcon build --packages-select autonomous_exploration
source install/setup.bash
ros2 launch autonomous_exploration autonomous_exploration.launch.py
```
